```python
# This is an example of a for loop
for name in names:
    print(name)
    #Condition statement block
    if name == 'buzz aldrion':
        print('Sup buzz\n')
        #Next value in the loop
        continue
    elif name == 'yuri gagarin':
        print('Hey yuri')
        # stops the loop
        break
    # Catch all
    else:
        # placeholder to do nothing
        pass
    #outside of if statement block
    print('Please to have you here!\n')
```

# 3. Organize All The Data in a Dictionary

```python
import pandas as pd
```

```python
# Create a dictionary with original name of league as key
dict_countries = {'Spanish La Liga' : 'SP1','Spanish Segunda Division' : 'SP2',
                 'German Bundesliga' : 'D1',
                 'English Premier League' : 'E0', 'English League 1': 'E1','English League 2': 'E2', 'English League 3' : 'E3'}
```

```python
# Get a dictionary element
dict_countries['Spanish La Liga']
```

```python
for league in dict_countries:
    print(league)
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data = {}
# Creating a list of leagues
# leagues = ['E0','E2','E3']


# Looping through range of seasons and insert new column
for league in dict_countries:
    frames = []
    for season in range(15,21):
        df = pd.read_csv(root + str(season) + str(season + 1)+ '/' + dict_countries[league] + '.csv')
        df.insert(1,'season',season)
        frames.append(df)  
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```

```python
# show dataframe inside dict_historical _data
dict_historical_data['English Premier League']
```
